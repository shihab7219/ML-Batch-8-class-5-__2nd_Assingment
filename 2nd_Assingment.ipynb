# Normalization

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
normalization_value = MinMaxScaler(feature_range=(1, 3))
```

```python
normalization_value_fit = normalization_value.fit(data[["Marketing Spend"]])
normalization_value_fit
```

```python
data["Marketing Spend"] = normalization_value.transform(data[["Marketing Spend"]])

data.head()
```

```python
normalization_value2 = MinMaxScaler()
data["Marketing Spend"] = normalization_value2.fit_transform(data[["Marketing Spend"]])
data.tail(2)
```

# Standardization

```python
from sklearn.preprocessing import StandardScaler
```

```python
standardization_value = StandardScaler()
```

```python
data["Administration"] = standardization_value.fit_transform(data[["Administration"]])
```

```python
data.tail(15)
```

# Max Absulate Scaler

```python
from sklearn.preprocessing import MaxAbsScaler
```

```python
max_abs_scaler_value = MaxAbsScaler()
```

```python
data["Transport"] = max_abs_scaler_value.fit_transform(data[["Transport"]])
```

```python
data.head(10)
```

# Robust Scaler

```python
from sklearn.preprocessing import RobustScaler
```

```python
robustscaler_value = RobustScaler()
```

```python
data["Profit"] = robustscaler_value.fit_transform(data[["Profit"]])
```
